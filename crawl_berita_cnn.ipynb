```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm.auto import tqdm
```

# Crawling Data
Teknik crawling data adalah proses pengumpulan informasi dari berbagai sumber di internet secara otomatis. Crawling data dilakukan oleh program yang disebut "web crawler". Web crawler pada proses ini menggunakan beautifulsoup untuk melakukan ekstraksi halaman konten tersebut agar dapat mengambil isi-isi konten pada berita di website CNN Indonesia. Terdapat juga libraly request yang digunakan untuk meminta respon pada situs yang akan dilakukan crawling data.


```python
from os import replace
def cnnnews(page, kategori, n, label):

    data = {'judul': [], 'berita': [], 'label': []}
    for i in tqdm(range(1, page+1)):
      url = f"https://www.cnnindonesia.com/{kategori}/indeks/{n}/{i}"
      r = requests.get(url)
      request = r.content
      soup = BeautifulSoup(request, 'html.parser')
      soup = soup.find('div', {'class': 'flex flex-col gap-5'})
      news = soup.findAll('article', {'class': 'flex-grow'})
      # news = soup.findAll('a', {'aria-label': 'link description'})

      for new in tqdm(news):
        a_element = new.find('a', {'aria-label': 'link description'})['href']
        detail_request = requests.get(a_element)
        detail_soup = BeautifulSoup(detail_request.content, 'html.parser')

        judul = detail_soup.find('h1', {'class': 'leading-9'})
        berita = detail_soup.find('div', {'class': 'detail-text'})

        if judul and berita:
          judul = judul.text
          berita = berita.text
          # noise = detail_soup.find('strong').text

          # berita = berita.replace("ADVERTISEMENT", "").replace("SCROLL TO CONTINUE WITH CONTENT", "").replace("Lihat Juga :", "").replace("Bagikan", "").replace(noise, "").replace("-- ", "").replace("--", "")
          berita = berita.replace("ADVERTISEMENT", "").replace("SCROLL TO CONTINUE WITH CONTENT", "").replace("Lihat Juga :", "").replace("Bagikan", "").replace("-- ", "").replace("--", "")
          berita = ' '.join(berita.split())

          label = label
          data["judul"].append(judul)
          data["berita"].append(berita)
          data["label"].append(label)

    # df = pd.DataFrame(data)
    # df.to_csv("berita-cnn.csv", index=False)

    return data
```

```python
ekonomi = cnnnews(30, 'ekonomi', 5, 1)
```

```python
olahraga = cnnnews(30, 'olahraga', 7, 2)
```

```python
edukasi = cnnnews(30, 'edukasi', 559, 3)
```

```python
hiburan = cnnnews(30, 'hiburan', 9, 4)
```

# Data to CSV

```python
ekonomi = pd.DataFrame(ekonomi)
olahraga = pd.DataFrame(olahraga)
edukasi = pd.DataFrame(edukasi)
hiburan = pd.DataFrame(hiburan)
# df.to_csv("berita-cnn.csv", index=False)
df = pd.concat([ekonomi, hiburan, edukasi, olahraga], ignore_index=True)
# df.to_csv("berita-cnn.csv", index=False)
```
